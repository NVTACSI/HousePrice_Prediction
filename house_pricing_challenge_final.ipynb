Now we compute the Skewness and Kurtosis; those are two measures that allow to evaluate the "shape" of a given distribution.

````python
# Print the values of skewness and kurtosis
print("Skewness: %f" % skew(trainDF['SalePrice']))
print("Kurtosis: %f" % kurtosis(trainDF['SalePrice']))
````

Skewness is usually described as a measure of a datasets symmetry. A perfectly symmetrical data set will have a skewness of 0. Generally, the normal distribution has a skewness of 0. The skewness value can be positive, negative, or undefined. 

Meanwhile, the coefficient of Kurtosis is a measure for the degree of tailedness in the variable distribution. Higher kurtosis means that more of the variance is the result of infrequent extreme deviations, as opposed to frequent modestly sized deviations. 

In this case (`SalePrice` variable), a value of ~2 for the Skewness and ~7 for the Kurtosis. +/- 2 is usually considered as an acceptable value for a normal distribution. However, a high value of Kurtosis indicates that a curve with an high pick and not exactly centered but slightly moved at left. Thus, both these measures suggest that the `SalePrice` variable is not normally distributed. As a further check for normality, we use the normal probability plot as following.


````python
# Probability plot (skewness)
plt.figure(figsize=(10, 6))
probplot(trainDF['SalePrice'], plot=plt)
plt.title("Probability plot of SalePrice", weight="semibold")
plt.grid()
plt.show()
````

In a normal probability plot (also called a "normal plot"), the sorted data are plotted vs. values selected to make the resulting image look close to a straight line if the data are approximately normally distributed. Deviations from a straight line suggest departures from normality. Therefore, the above figure once again confirms that the sale prices are not follow normal distribution.

With the previous plots, e can appreciate the skewness of the data and notice that we have a left-skewed target variable. It is normal in reality (only few people have an expensive house) but it's not good for our models.

In particular, in the case of our linear regression problem, where we try to minimize the square of the losses: using original target values we will have huge errors on the most expensive houses, and for this reason the algorithm leads to fit better these points in order to minimize the overall error. As a consequence, he expensive houses will drive our model and the cheapest ones will be on the back burner.

In partice,  one of noticeable things is that most of algorithms works better with normal distribution. Therefore, it is a good idea to normalize the `SalePrice` features before using for training. We will have more detailed analysis about this idea in the part of `Feature Transformation`. 

To finish with the analysis on `SalePrice` variable, we also visualize two additional plots: an ECDF and a box plot.

````python
def plot_ECDF(data, title, xlabel, scale='linear'):
    """Plot the empirically cumulative distribution function given a data.
    
    Parameters
    ----------
    data: numpy array
        The data used to plot ECDF.
    title: string
        The title of the plot.
    xlabel: string
        The label for x-axis
    scale: string
        The name of scaling function applied to the data.
    """
    # Create a plot
    plt.figure(figsize = (10, 6))
    plt.plot(data, np.arange(len(data))/float(len(data)), linewidth=2)
    
    # Label the plot
    plt.yticks(np.arange(1.1, step=0.1))
    plt.xlabel(xlabel)
    plt.ylabel('ECDF')
    plt.grid(True, which="both", ls="-")
    plt.title(title, weight="semibold")
    plt.xscale(scale)


def annotate_percentiles(data, percentiles=[0.10, 0.25, 0.50, 0.75, 0.90]):
    """Annotates percentiles over a plot of ECDF.
    
    Parameters:
    data: numpy array
        The input data 
    percentiles: list of float
        The list contains the considered percentiles.
    """
    for y in percentiles:
        x = data[int(len(data)*y)]
        plt.vlines(x, 0, y, linestyle="dashed", alpha=0.8, colors='g')
        plt.hlines(y, 0, x, linestyle="dashed", alpha=0.8, colors='g')
        plt.annotate('%s' % x , xy=(x, y), )
        plt.scatter(x, y, zorder=2, color="red")
````

````python
# Plot the ECDF of the `SalePrice`
prices = np.sort(trainDF.SalePrice.tolist())
plot_ECDF(prices, 'ECDF of Sale Prices', 'Sale Prices')
annotate_percentiles(prices)
````

````python
# Show the box plot of the SalePrice
plt.figure(figsize = (14, 4))
plt.boxplot(trainDF.SalePrice, vert=False, patch_artist=True)
plt.title('Sale Price box plot', weight="semibold")
plt.grid()
plt.show()
````

The ECDF is also useful to eventually show the presence of outliers: all the values which lies on the upper (or lower) part of the curve (when the trend stabilizes at 0 or 1) can be considered outliers.

The box plot diagram, in fact, helps in defining the upper limit and lower limit beyond which any data lying will be considered as outliers, showing them in a more direct and immediate way with respect to the ECDF. The main purpose of this diagram is to identify outliers before making any further observation so that the conclusion made from the study gives more accurate results not influenced by any extremes or abnormal values. Therefore, it's easy to see how the box plot marks the lower quartile, the upper quartile, and the lower and upper limits. So any value that will be more than the upper limit or lesser than the lower limit could be an outlier. Only the data that lies within lower and upper limit are statistically considered normal.

In our case, the plots suggest that some of the highest values in the dataset can be outliers. ### 1.3.1 Types of Features

Before starting visually analyze the features, we should divide them into suitable sub-groups. Basically, there are three main sub-groups.

* **Categorical** feature: (sometimes called a nominal feature) is one that has two or more categories, but there is no intrinsic ordering to the categories.  For example, gender is a categorical variable having two categories (male and female) and there is no intrinsic ordering to the categories.
* **Ordinal** feature: is similar to a categorical variable.  The difference between the two is that there is a clear ordering of the variables.  For example, suppose you have a variable, house quality, with three categories (low, medium and high).
* **Numerical** feature: includes oth interval features, ordinal variables that hold a measure of interval between items but have no absolute zero (e.g. the date of construction), and ratio features, like interval ones, but they can be compared using ratios (e.g. the size of the garage/pool).

### 1.3.2 Categorical Features

In order to have better insights about the distributions of categorical or ordinal features, we use a count plot together with a violin plot for each feature. The violin plot allows us to have the view of overall distribution of each feature with respect to the sale price. This may lead to some observation about the correlation between each feature to the sale price. Because we reuse these plots for categorical and oridinal features, we define a wrapper function for later reuse.

````python
def plot_count_violin(featuresDF, targetDF):
    """Plot count plot and violin plot for categorical or ordinal features.
    
    Parameters
    ----------
    featureDF: pandas DataFrame
        The dataframe contains the set of features
    targetDF: pandas DataFrame
        The dataframe contains the corresponding target (saleprice)
    """
    # Visualize the distribution of categorical features
    for featureName in featuresDF.columns:
        _, ax = plt.subplots(1, 2, figsize=(15, 3))

        # Plot the count plot
        sns.countplot(featuresDF[featureName], ax=ax[0], alpha=0.7)
        ax[0].grid()
        ax[0].set_title("Count plot")

        # Plot the strip plot
        sns.violinplot(x=featuresDF[featureName], y=targetDF.SalePrice,  ax=ax[1])
        
        ax[1].grid()
        ax[1].set_title("Violin plot")

        # Adjust the orientation of the names in x-axis to avoid overlapping
        if len(plt.xticks()[1]) >= 5:
            ax[0].set_xticklabels(labels=plt.xticks()[1], rotation=70)
            ax[1].set_xticklabels(labels=plt.xticks()[1], rotation=70)

        # Add the title
        plt.suptitle(featureName, weight="semibold")
````

As can be seen from the above plot, the number of missing values of `MasVnrType` and `MasVnrArea` features are the same. Let's print the rows of data containing the corresponding missing values. Notice that we will repeat this step for other group of features containing missing values. Therefore, we should define a wrapper function for later uses.

````python
def get_missing_rows(dataFrame, testFeatures, not_missing=False, displayedCols=None):
    """Get the rows containing non-missing or missing values of a set of features.
    
    Parameters
    ----------
    dataFrame: pandas DataFrame
        The input data frame
    testFeatures: list of string
        The list containing the names of features to be 
    not_missing: boolean
        The variable indicates whether we get the rows containg missing or 
        non-missing values.
    displauedCols: list of string
        The list containing the columns that will be shown
        
    Parameters:
    resultDF: pandas DataFrame
        The resulting data frame.
    """
    # Get the indexes of rows containing missing values
    idxes = None
    for featureName in testFeatures:
        if not_missing == False:
            idx = dataFrame[featureName].isnull()
        else:
            idx = dataFrame[featureName].notnull()
            
        if idxes is None:
            idxes = idx
        else:
            if not_missing == False:
                idxes = (idxes | idx)
            else:
                idxes = (idxes & idx)
    
    # Return the results
    resultDF = dataFrame[idxes]
    if displayedCols is not None:
        resultDF = resultDF[displayedCols]

    return resultDF
````

## 2.2 Feature Encoding

As can be seen from the previous sections, the types of categorical or ordinal features sometimes are string. Meanwhile, the regression models are usually built on numerical values. Therefore, it is necessary to encode the categorical and ordinal features properly.
Obviously, in order to make the model generalize, all the data that we will input must respect the same encoding, so we will encode in the same way the test data.

First of all, lets create copies of clean data frames for feature-encoding steps. This helps guarantee that we can rollback in the cases of error happening.

````python
# Create a copy of clean data frames for encoding procedure
encodedTrainDF = cleanTrainDF.copy(deep=True)
encodedTestDF = cleanTestDF.copy(deep=True)
````

### 2.2.1 Ordinal Features

In order to encode the ordinal features, we could use the `LabelEncoder` utility provdied by scikit-learn package. A benefit of this approach is its simplicity since we don't have to define a mapping from string to integer for each ordinal feature. However, it doesn't keep the order of categories of each feature. For example, in the case of `BsmtQual` feature, the posibilities of values are `Ex` (Excellent), `Gd` (Good), `TA` (Typical), `Fa` (Fair), `Po` (Poor) and `NA` (No basement). ## 2.3 Outlier Detection

Outliers are extreme values that deviate from other observations on data , they may indicate a variability in a measurement, experimental errors or a novelty. In other words, an outlier is an observation that diverges from an overall pattern on a sample. There are a variety of reasons that cause outliers on a data set such as human errors, instrument errors, Data processing errors, etc. Detecting outliers is of major importance for machine learning since the quality of data is as important as the quality of a prediction or classification model. 

In machine learning and data analytics, clustering methods are useful tools that help us visualize and understand data better. Relationships between features, trends and populations in a data set can be graphically represented via clustering methods like DBSCAN, and can also be applied to detect outliers in nonparametric distributions in many dimensions. DBSCAN is a density based clustering algorithm, it is focused on finding neighbors by density (`MinPts`) on an n-dimensional sphere with radius $\epsilon$. A cluster can be defined as the maximal set of density connected points in the feature space.

Although DBSCAN can work well when dealing with higher dimensional feature spaces 4 or more dimensions, However, in our given dataset, it includes lots of data types: categorical, numerical, ordinary that makes hard to choose a good neighborhood distance (defined by $\epsilon$). Then one possible way is to detect latent dimensions in our dataset in order to tranform our original dataset into a new  feature space that can easily apply DBSCAN algorithm to dectect outlier. In order to achieve that, we can use dimensional reduction techniques and Principal Component Analysis (PCA) is an useful one. PCA is a linear dimensionality reduction technique using Singular Value Decomposition of the data to project it to a lower dimensional space. Then we need to define how many dimensions we should keep in the new feature space while holding enough information (based on the eigen-values).

````python
# using PCA transformation to create new feature space
pca = PCA(n_components=2)
PCATrain = pca.fit_transform(encodedTrainDF.copy(deep=True))
pca.explained_variance_ratio_
````

````python
# plot the new feature space after applying PCA
plt.figure(figsize=(8,5))
ax  = sns.scatterplot(x=PCATrain[:,0], y=PCATrain[:,1], s=70)
ax.grid()
ax.set_xlabel('Attr0', color='r', size=12)
ax.set_ylabel('Attr1', color='r', size=12)
ax.set_title('New feature space after PCA', color='r', size=15)
plt.show()
````

As we can see, retaining 2 dimensions in the new space can keep almost information. Now, it is time to look how well DBSCAN algorithm work on our new feature space. The first step is scaling the data, since the radius $\epsilon$ will define the neighborhoods along with MinPts. After scaling the feature space, is time to choose the spatial metric on which dbscan will perform the clustering. The metric must be chosen depending on the problem. Then, the parameter eps ($\epsilon$) must be chosen accordingly to perform clustering. If $\epsilon$ is too big many points will be density connected, if its too small the clustering will result in many meaningless clusters. DBSCAN is also sensitive to the `MinPts` parameter, tuning it will completely depend on the problem at hand. In the notebook, we only show the best obtained parameter after tuning.

````python
# scale the given dataset by RobustScaler
scalePCATrainDF = pd.DataFrame()
scaler = RobustScaler(quantile_range=(0.0, 100.0))
for i in range(PCATrain.shape[1]):
    x = PCATrain[:,i].reshape(-1, 1) #returns a numpy array
    x_scaled = scaler.fit_transform(x)
    scalePCATrainDF['attr'+str(i)] = x_scaled[:,0]

# applying DBSCAN technique to dectect outliers in given dataset
clustering = DBSCAN(eps=0.18, min_samples=15, metric='manhattan').fit(scalePCATrainDF)
outliersIdxs = scalePCATrainDF[(clustering.labels_ == -1)].index
scalePCATrainDF['isOutlier'] = scalePCATrainDF.index.isin(outliersIdxs)

# plot the figure to show the outlier's location in the given dataset
plt.figure(figsize=(8,5))
ax  = sns.scatterplot(x='attr0', y='attr1', s=70, hue='isOutlier', data=scalePCATrainDF)
ax.grid()
ax.set_xlabel('Attr0', color='r', size=12)
ax.set_ylabel('Attr1', color='r', size=12)
ax.set_title('New feature space after PCA', color='r', size=15)
plt.show()
````

### 2.4.1 Feature Extraction

After having deeply analyzed the dataframe attributes and their values, we want to expand the original attribute-set by creating more features that can be used for giving out better predictions. We do that manually, by trying to create new features by simple operators on the existing features.

````python
def addFeature_elementwiseSum(pandas_df, attributes, new_attr):
    """Create a new feature by elementwise sum from list of features
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame
    attributes: list of string
        List of features used to create a new feature
    new_attr: string
        The name of the new feature
        
    Returns
    -------
    pands_df: pandas DataFrame
        The output data frame
    """
    pandas_df[new_attr] = pandas_df[attributes[0]].values
    for attr in attributes[1:]:
        pandas_df[new_attr] += pandas_df[attr].values
        
    return pandas_df

def addFeature_bathCount(pandas_df):
    """Create a new feature representing the total number of bath of a house.
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame
        
    Returns
    -------
    pands_df: pandas DataFrame
        The output data frame
    """
    pandas_df['TotBaths'] = (pandas_df['FullBath']+pandas_df['BsmtFullBath']) + 0.5*(pandas_df['HalfBath']+pandas_df['BsmtHalfBath'])
    return pandas_df

def addFeature_boolean(pandas_df, attribute, new_attribute):
    """Create a boolean feature indidicates the existence of a feature.
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame
    attributes: string
        The name of the feature used to create a new feature
    new_attr: string
        The name of the new feature
        
    Returns
    -------
    pands_df: pandas DataFrame
        The output data frame
    """
    pandas_df[new_attribute] = pandas_df[attribute].apply(lambda x: 1 if x > 0 else 0)
    
    return pandas_df

def add_features(pandas_df):    
    """Execute the steps of adding features on a given data frame.
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame
        
    Returns
    -------
    pands_df: pandas DataFrame
        The output data frame
    """
    # TotSF
    pandas_df = addFeature_elementwiseSum(pandas_df, ['1stFlrSF', '2ndFlrSF', 'BsmtFinSF1', 'BsmtFinSF2'], 'TotSF')
    
    # HasGarage
    pandas_df = addFeature_boolean(pandas_df, 'GarageArea', 'HasGarage')
    
    # HasBsmt
    pandas_df = addFeature_boolean(pandas_df, 'TotalBsmtSF', 'HasBsmt')
    
    # HasFirePlace
    pandas_df = addFeature_boolean(pandas_df, 'Fireplaces', 'HasFireplace')
    
    # BathCount
    pandas_df = addFeature_bathCount(pandas_df)
    
    # TotPorchSF
    pandas_df = addFeature_elementwiseSum(pandas_df, ['OpenPorchSF', 'EnclosedPorch', '3SsnPorch', 
                                                      'ScreenPorch', 'WoodDeckSF'], 'TotPorchSF')
    # HasPool
    pandas_df = addFeature_boolean(pandas_df, 'PoolArea', 'HasPool')
    
    # Has2ndFlr
    pandas_df = addFeature_boolean(pandas_df, '2ndFlrSF', 'Has2ndFlr')
    
    return pandas_df
````

### 2.4.2 Feature Transformation

Another important milestone for this work is the feature transformation. One of the first things we have seen in the first section is the SalePrice distribution which is right skewed, and we know that the existing linear models work better with a normal distributions. So we need to transform this variable and make it more normally distributed.

In this regard, we can try to analyze the distribution of all the other attributes, and eventually adjust it to simplify the models lives.

For the transformation we will use the np.log and the boxcox functions. We cannot use only the log transform because among the considered attributes there are some which contains zeroed values and for them the simple log cannot be applied. Anyhow, the boxcox function has a $\lambda$ parameter, which if is set to 0, make it assume the same behaviour of the log.

$$
y(\lambda)
\begin{cases}
\frac{y^{\lambda}-1}{\lambda}, \quad if \quad \lambda \neq 0;\\
\log(y), \quad if \quad \lambda = 0.
\end{cases}
$$

We will use then the boxcox_normmax function to estimate the best $\lambda$ value for the given distribution.

````python
# Create copies version of encoded data to use for transformation steps
transformedTrainDF = encodedTrainDF.copy()
transformedTestDF = encodedTestDF.copy()
````

````python
def transform_feature(pandas_df, feature, transformation=np.log, lambda_=None):
    """Apply transformation on a feature.
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame
    feature: string
        The name of the feature
    transformation: function
        The transformation function that will be applied on the feature
    lambda_: floeat
        The lambda value of cobox function
    
    Returns
    -------
    pandas_df: pandas DataFrame
        The output data frame
    -------
    """
    if(not lambda_):
        pandas_df[feature] = transformation(pandas_df[feature])
    else:
        pandas_df[feature] = transformation(pandas_df[feature], lambda_)
        
    return pandas_df
````

````python
def plot_distribution(values, attribute):
    """Utility function to plot the distribution and the probability plots given the values array
    
    Parameters
    ----------
    values: np.array or pandas series/dataframe
        The input values array
    attribute: string
        The name of the feature coresponding to the values array
    """
    print("Skewness: %f" % skew(values))
    print("Kurtosis: %f" % kurtosis(values))
    
    _, ax = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot the distribution
    sns.distplot(values, ax=ax[0], fit=stats.norm)
    ax[0].grid()
    ax[0].set_title(attribute +" distribution")

    stats.probplot(values, plot=ax[1])
    ax[1].grid()
    ax[1].set_title(attribute + " probability plot against quantiles of normal distribution")
    
    plt.tight_layout()
    plt.show()
````

First of all, we should define a utility function used to transform the skewed features of a given dataset. 

````python
def transform_skews(pandas_df, skew_map, skewness_limit=0.75):
    """Transform skewed features by using boxcox function.
    
    Parameters
    ----------
    pandas_df: pandas DataFrame
        The input data frame.
    skew_map: dictionary
        The dictionary contains the degree of skewness of each feature.
    skewness_limit: float
        The skewness threshold of applying boxcox transformation.
    
    Returns
    -------
    pandas_df: pandas DataFrame
        The resulting data frame
    """
    for attr, skew in skew_map.iteritems():
        if(math.fabs(skew) > skewness_limit):
            print("Applying transformation to", attr, ". Degree of skewness:", skew)
            pandas_df = transform_feature(pandas_df, attr, transformation=boxcox1p,
                                          lambda_=boxcox_normmax(pandas_df[attr]+1))
    return pandas_df
````

### 2.4.4 Feature Selection

Feature Selection is one of the core concepts in machine learning which hugely impacts the model performance. This step help to identify the related features from a set of data and removing the irrelevant or less important features with do not contribute much to our target variable. Therefore, performing feature selection before modeling data will help to reduce overfitting,  achieve better accuracy as well as reduce training time. 
There are lots of criterias to perform feature selection. In this notebook, we show our work on 2 feature selection techniques:Univariate Selection (Mutual Information) and Feature Importance (Lasso regularization)

** Mutual information **

Mutual information is a measure between two (possibly multi-dimensional) random variables X and Y, that quantifies the amount of information obtained about one random variable, through the other random variable. The mutual information is given by

$$ I(X,Y) = \int_{X} \int_{Y} p(x,y) \log \frac{p(x,y)}{p(x)p(y)}$$

where $p(x,y)$ is the joint probability density function of $X$ and $Y$, and where $p(x)$ and $p(y)$ are the marginal density functions. The mutual information determines how similar the joint distribution $p(x,y)$ is to the products of the factored marginal distributions. If $X$ and $Y$ are completely unrelated (and therefore independent), then $p(x,y)$ would equal $p(x)p(y)$, and this integral would be zero. In general, the higher the mutual information value between two random variables X and Y, the more related between them. Now, in term of model selection, we would like to keep features that are highly related to our target variable and remove nearly independent ones. In other words, we can select our features from feature space by ranking their mutual information with the target variable.

** Lasso regularization **

As described above, mutual information is a good criterion to see how independent between 2 random variables and then can be useful to do feature selection. However, mutual information is usually considered as univariate feature selection metric where it only reflects the effect of a single input feature to target while we also need to analyze the correlation between input features to each other. The reason is features that are highly correlated or co-linear (aka multicollinear phenomenon) can cause model overfitting. When a pair of variables are highly correlated we can remove one to reduce dimensionality without much loss of information. There are a variety of methods that help to do multivariate feature selection: wrapper, filter, embedded, etc. In this notebook, we would like to show our work on Lasso Regularization method. This is a regression technique that not only try to minimize the loss function (i.e: Mean Squared Error) but also add a constraint on the size of the model's coefficients $w$:

$$ RSS + \lambda \left( \sum_{j=1}^p \mid{w_{j}}\mid  \right) $$

Here, $\lambda > 0$  is a complexity parameter that controls the amount of coefficient shrinkage: the larger the value of $\lambda$, the greater the amount of shrinkage and thus the coefficients become more robust to collinearity. That is why Lasso regression does a form of parameter selections since the feature variables that arent selected will have a total weight of 0.

````python
# apply Lasso model on the given dataset to do feature selection
lassoModel = Lasso(alpha=0.0005)
lassoModel.fit(Xtrain, ytrain)
lassoCoefDF = pd.DataFrame({"attr": Xtrain.columns,
                    "value": np.abs(lassoModel.coef_)})

print("The number of attribute having zero coefficent is", sum(lassoCoefDF.value == 0))

print("Some atributes having zero coefficent:")
display(lassoCoefDF.sort_values(["value", "attr"], ascending=True).head(20))
````

** Ridge **<br> 

Traditional linear fitting involves minimizing the RSS (residual sum of squares) in which multicollinearity can cause overfitting phenomenol. Ridge regression addresses this problem by imposing a penalty on the size of coefficients. In ridge regression, a new parameter is added, and now the parameters will minimize:<br> 

$$ RSS + \lambda \left( \sum_{j=1}^p w_{j}^2  \right) $$

Here, $\lambda > 0$  is a complexity parameter that controls the amount of shrinkage: the larger the value of $\lambda$, the greater the amount of shrinkage and thus the coefficients become more robust to collinearity. This parameter is tuned using cross-validation as it must minimize the test error. Therefore, a range of lambdas is used to fit the model and the lambda that minimizes the test error is the optimal value.<br> 
<br> 
** Lasso **<br> 

Similarly to ridge regression, lasso will minimizes:

$$ RSS + \lambda \left( \sum_{j=1}^p \mid{w_{j}}\mid  \right) $$

Notice that we use the absolute value of the parameter $w$ instead of its squared value. Also, the same tuning parameter is present. However, if lambda is large enough, some coefficients will effectively be 0. The Lasso is prefered in case we want to estimate sparse coefficients in our model. It is useful in some contexts due to its tendency to prefer solutions with fewer parameter values, effectively reducing the number of variables upon which the given solution is dependent.<br> 
<br> 
** ElasticNet **<br> 

ElasticNet is a hybrid of Lasso and Ridge Regression techniques. It is uses both the `L1` and `L2` regularization taking on the effects of both techniques: <br> 
<br> 
$$ RSS + \lambda_1 \left( \sum_{j=1}^p w_{j}^2  \right) + \lambda_2 \left( \sum_{j=1}^p \mid{w_{j}}\mid  \right) $$
<br> 
This combination allows for learning a sparse model where few of the weights are non-zero like Lasso, while still maintaining the regularization properties of Ridge.<br> 
<br> 
** Bayesian Ridge **<br> 

The problem solved by Ridge regression can be similarly treated as Bayesian method. Bayesian regression techniques can be used to include regularization parameters in the estimation procedure: the regularization parameter is not set in a hard sense but tuned to the data at hand. This can be done by introducing uninformative priors over the hyper parameters of the model. The 
regularization $\lambda$ used in Ridge Regression is equivalent to finding a maximum a posteriori estimation under a Gaussian prior over the parameters with precision $\alpha^{-1}$. Instead of setting $\lambda$ manually, it is possible to treat it as a random variable to be estimated from the data.

To obtain a fully probabilistic model, the output $y$ is assumed to be Gaussian distributed around $Xw$:<br> 
<br> 
$$ p(y \mid X, w, \lambda) = \mathcal{N}(y \mid Xw, \lambda)$$
<br> 
and the prior for the parameter $w$ is given by a spherical Gaussian: 
<br> 
$$ p(w \mid \alpha) = \mathcal{N}(w \mid 0,  \alpha^{-1}I_p)$$
<br> 
The priors over $\alpha$ and $\lambda$ are chosen to be gamma distributions, the conjugate prior for the precision of the Gaussian. The resulting model is called Bayesian Ridge Regression, and is similar to the classical Ridge. The parameters $w$, $\alpha$ and $\lambda$ are estimated jointly during the fit of the model. The remaining hyperparameters are the parameters of the gamma priors over $\alpha$ and $\lambda$. These are usually chosen to be non-informative. The parameters are estimated by maximizing the marginal log likelihood.

In random forests, each tree in the ensemble is built from a sample drawn with replacement (i.e., a bootstrap sample) from the training set. Random forest can be applied in both classification and regression problem. In case of regression, Mean Squared Error (MSE) or Mean Absolute Error (MAE) is the criterion to choose features to split at each non-leaf node. ** Gaussian Process **
<br> 
Gaussian Processes (GP) are a generic supervised learning method designed to solve regression and probabilistic classification problems. Gaussian Processes begin with we looking for a model for our problem but we do not know how exactly it is. We first define a probability function to represent our belief about the desired fucntion (aka prior function). Then, we will collect samples (training data $F_N, X_N$) to update our belief which lead to produce posterior function. <br> 
<br> 

$$ p(F_N \mid X_N) = \frac{exp(-\frac{1}{2}F_N^TK_N^-1F_N)}{\sqrt{(2\pi)^Ndet(K_N)}}$$ <br>

where $K_N$ is a NxN covariance matrix and its coefficients are expressed in terms of a correlation function (or kernel) $K_{mn} = K(x_m,x_n,\theta)$. The hyper- parameters $\theta$ of the kernel are calibrated according to a maximum likelihood principle. $K_N$ is chosen to reflect a prior assumption of the function and therefore the choice of the kernel will have a significant impact on the correctness of the regression.<br>

Hence, GP describes probability distributions over functions by using Bayes rule to update our distribution of functions by observing training data. The updated Gaussian process is constrained to the possible functions that fit our training data that also mean it narrows the range of possible fuctions for our problem.

# 4. Parameter Optimization

## 4.1 Grid Search Optimization 

Now, we might be interested in selecting the best hyperparameters for a selected machine learning method. A model hyperparameter is a characteristic of a model that is external to the model and whose value cannot be estimated from data. The value of the hyperparameter has to be set before the learning process begins. For example, c in Support Vector Machines, the number of hidden layers in Neural Networks. There are many techniques that help to tune the optimal hyperparameter for a selected model. We firstly introduce "Grid search" method to select the best among a defined range of parameters for each model.

The implementation of Grid search technique is defined as a generic wrapper blow class for convenience of use (suports a wide-range of machine learning models.

````python
class GridSearchOptimizer():
    """A generic wrapper class for Grid Search Optimization."""
    def __init__(self, model, params, X, y, cv=10, n_jobs=1, verbose=1):
        """Intialization.
        
        Parameters
        ----------
        model: sckit-learn/xgboost models.
            The machine learning model needed to be optimized
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        cv: int
            Number of cross-validation folds.
        n_jobs: int
            Number of parallel processes.
        verbose: int
            Indicate whether to print the trace of optimization.
        """
        self.model = model
        self.params = params
        self.X = X
        self.y = y
        self.cv = cv
        self.n_jobs = n_jobs
        self.optimizer = None
        self.verbose = verbose

    
    def optimize(self):
        """Optimize the hyper-paremeter by Grid Search treatment.
        
        """
    
        self.optimizer = GridSearchCV(self.model, self.params, cv=self.cv, n_jobs=self.n_jobs, 
                              scoring="neg_mean_squared_error", verbose=self.verbose)

        self.optimizer.fit(self.X, self.y)
        
    def get_best(self):
        """Get the best optmized hyper-parameters.
        
        Return
        ------
        best_score: float
            The best rmse score
        best_params: dict
            The best hyper-parameters
        """
        best_score = np.sqrt(-self.optimizer.best_score_)
        best_params = self.optimizer.best_params_
        return best_score, best_params
````

## 4.2 Bayesian Optimization

A typical approach for tuning parameters of an online system is to manually run small grid searches to optimize each parameter separately. Bayesian optimization constructs a statistical model of the relationship between the parameters and the online outcomes of interest, and uses that model to decide which experiments to run. This model-based approach has several key advantages, especially for tuning online machine learning systems.

* **Better scaling with parameter dimensional**ity: Given the limitations on the number of online experiments that can be run, grid search or manual tuning cannot be used to tune more than a couple parameters simultaneously. The use of models allows Bayesian optimization to scale to a much larger number of parameters; we routinely optimize up to 20 parameters jointly. This is important for machine learning systems where there are frequently interactions between parameters that necessitate joint optimization.

* **Reduced number of experiments**: Bayesian optimization allows us to borrow strength across rounds of experimentation: A test of a parameter value in a continuous space gives us information about not only its outcome, but also those of nearby points. This allows us to greatly reduce the number of experiments needed to explore the space.

* **Better experiment outcomes**: The model is able to identify parts of the space that are unlikely to provide a good outcome, and avoid running tests of those parameter values. This improves the experience inside the treatment groups.

* **Understand the parameter space**: Modeling allows us to visualize and better understand how the parameters affect the outcome of interest. This plot shows a 2-d slice of an 8 parameter experiment, and is typical of the visualizations we make to better understand the parameter space:

<p align="center"> 
    <img src="https://i.ibb.co/kGC5281/contour-opt.png">
</p>

Bayesian optimization is a technique for solving optimization problems where the objective function (i.e., the online metric of interest) does not have an analytic expression, rather it can only be evaluated through some time consuming operation (i.e., a randomized experiment). The key to efficiently exploring a multidimensional space with a limited number of experiments is modeling: The true objective function is unknown, but we can fit a model to the observations we have so far and use the model to predict good parts of the parameter space where we should run additional experiments.

The Gaussian process (GP) is a nonparametric Bayesian model that works well for Bayesian optimization because it provides excellent uncertainty estimates and is analytically tractable. It provides an estimate of how an online metric varies with the parameters of interest, as illustrated here:

<p align="center"> 
    <img src="https://i.ibb.co/TLnQY0V/gp-opt.png">
</p>

Each data marker in the figure above corresponds to the outcome of a blackbox function of that parameter value. We can use the GP to decide which parameter to test next by balancing exploration (high uncertainty) with exploitation (good model estimate). This is done by computing an acquisition function that estimates the value of running an experiment with any given parameter value.

In order to take advantages of this method, we use the package `BayesianOptimization` (https://github.com/fmfn/BayesianOptimization). We create a generic wrapper class for convenience of use (suports a wide-range of machine learning models). A machine leaning model is treated as a blackbox function. We then try to optimize this function, which corresponds to finding best hyper-parameters of the machine learning model, by using Bayesian Optimization.

````python
class BayesianOptimizer():
    """A generic wrapper class for Bayesian Optimization."""
    def __init__(self, model, X, y, cv=10, n_jobs=1, verbose=1):
        """Intialization.
        
        Parameters
        ----------
        model: sckit-learn/xgboost models.
            The machine learning model needed to be optimized
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        cv: int
            Number of cross-validation folds.
        n_jobs: int
            Number of parallel processes.
        verbose: int
            Indicate whether to print the trace of optimization.
        """
        self.model = model
        self.X = X
        self.y = y
        self.cv = cv
        self.n_jobs = n_jobs
        self.optimizer = None
        self.verbose = verbose
        
    def blackbox_fn(self, **kwargs):
        """A generic blackbox function need to be estimated.
        
        This function allows different sets of hyper-parameters for each model.
        """
        # Get the hyper-parameters from the arguments
        params = {}
        for key, value in kwargs.items():
            if key in ["max_depth", "n_estimators"]:
                params[key] = [value.astype(int)]
            elif key in ["hidden_layer_sizes"]:
                params[key] = [[value.astype(int)]]
            else:
                params[key] = [value]
        
        # This is quite tricky for the generality purporse
        # We use GridSearchCV functions from sckit-learn that allows
        # pass a dictionary of parameters into the model
        # Notice that we just fit the model one time (grid search space is only 1 point)
        gridSearch = GridSearchCV(self.model, params, cv=10, n_jobs=1, 
                                  scoring="neg_mean_squared_error")
        gridSearch.fit(self.X, self.y)
        score = np.sqrt(-gridSearch.best_score_)

        return -score
    
    def optimize(self, pbounds, init_points=10, n_iter=10):
        """Optimize the hyper-paremeter by Bayesian treatment.
        
        Parameters
        ----------
        pbounds: dict
            A dictionary contains the lower and upper bounds for each hyper-parameter
        init_points: int
            The number of points used for initialization.
        n_iter: int
            The number of iterations used for optimization.
        """
        self.optimizer = BayesianOptimization(f=self.blackbox_fn, pbounds=pbounds, verbose=self.verbose)
        self.optimizer.maximize(init_points=init_points, n_iter=n_iter)
        
    def get_best(self):
        """Get the best optmized hyper-parameters.
        
        Return
        ------
        best_score: float
            The best rmse score
        best_params: dict
            The best hyper-parameters
        """
        best_score = -self.optimizer.max['target']
        best_params = self.optimizer.max['params']
        return best_score, best_params
````

````python
# Create X, y dataset from the training data in which the features are selected
X = selectedTrainDF.copy().drop(['SalePrice'], axis=1)
y = selectedTrainDF.copy().SalePrice

# Loop over the models and conduct Bayesian Optimization
for modelName in bayesOptimModels.keys():
    print("-"*20 + "\n {}\n".format(modelName) + "-"*20)
    
    # Define the Bayesian optimizer
    model = bayesOptimModels[modelName]
    optimizer = BayesianOptimizer(model['model'], X, y, cv=10, n_jobs=-1)
    
    # Optimize hyper-parameters and printout the result
    optimizer.optimize(model['pbounds'], n_iter=10)
    best_score, best_params = optimizer.get_best()
    
    # Save the result
    model['best_score'] = best_score
    model['best_params'] = best_params
    
    print("Best score: ", best_score)
    print("Best hyper-parameters: ", best_params)
    print("="*50)
````

## 5.2 Ensemble of Models

Ensemble methods are commonly used to boost predictive accuracy by combining the predictions of multiple machine learning models. The traditional wisdom has been to combine so-called weak learners. However, a more modern approach is to create an ensemble of a well-chosen collection of strong yet diverse models.

### 5.2.1 Models Averaging

The simplest kind of ensemble is the unweighted average of the predictions of the models that form a model library. For example, if a model library includes three models for an interval target (as shown in the following figure), the unweighted average would entail dividing the sum of the predicted values of the three candidate models by three. In an unweighted average, each model takes the same weight when an ensemble model is built.

<p align="center"> 
    <img src="https://i.ibb.co/Wskvyjr/weighted-unweighted.png">
</p>

Lets define a class supporting ensemble learning by averaging multiple models.

````python
class AveragingModels(BaseEstimator, RegressorMixin, TransformerMixin):
    """The class representing a averaging model from a set of models."""
    def __init__(self, models):
        """Initialization.
        
        Parameters
        ----------
        models: list 
            List of models used to build the averaging model.
        """
        self.models = models
        self.weights = np.ones(len(models))
        
    def fit(self, X, y):
        """Fit the data on clones of the original models.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        """
        # We define clones of the original models to fit the data in
        self.models = [clone(x) for x in self.models]
        
        # Train cloned base models
        for model in self.models:
            model.fit(X, y)

        return self
    
    def predict(self, X):
        """Do the predictions of all base models on the test data.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the testing data
            
        Returns:
        predictions: 1D numpy array
            The predictions by the averaged model.
        """
        # use the averaged predictions as meta-features for the final prediction 
        # which is done by the meta-model
        predictions = np.column_stack([
            model.predict(X) for model in self.models
        ])
        
        predictions = np.dot(predictions, self.weights/sum(self.weights))  
        
        return predictions
    
    def get_cv_score(self, X, y, cv=10):
        """Compute the k-fold validation on a given data.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        cv: int
            The number of folds used to test
        
        Returns
        scoreMean: float
            The average of mean-squared errors
        scoreStd: float
            The standard deviation of mean-squared errors
        -------
        """
        scores = np.sqrt(-cross_val_score(self, X, y, scoring='neg_mean_squared_error', cv=cv))
        return scores.mean(), scores.std()
````

First of all, lets create the top models using hyper-parameter obtained by Bayesian optimiazation.

````python
# Ridge
ridgeModel = Ridge(**bayesOptimModels["Ridge"]["best_params"], random_state=2019)

# ElasticNet
elasticNetModel = ElasticNet(**bayesOptimModels["ElasticNet"]["best_params"], random_state=2019)

# Bayesian Ridge
bayesianRidgeModel =  BayesianRidge(**bayesOptimModels["BayesianRidge"]["best_params"])

# Gaussian Processes
gpModel = GaussianProcessRegressor(DotProduct() + WhiteKernel(), **bayesOptimModels["GaussianProcessRegressor"]["best_params"], random_state=2019)
````

Next, define the averaging model using the above element models.

````python
# Create the ensemble model
models = [ridgeModel, bayesianRidgeModel, gpModel]
averaged_models = AveragingModels(models=models)

# Evaluate the ensemble model using 10-fold cross validation
score_mean, score_std = averaged_models.get_cv_score(X, y, 10)
print("RMSE of the averaged model: {:.4f} , std: {:.4f}".format(score_mean, score_std))
````

As expected, by taking advantages of many predictive models, the performance of the averaged model is higher than those of the based models.

### 5.2.2 Models Stacking

More generally, we can think about using weighted averages. For example, we might believe that some of the models are better or more accurate and we want to manually assign higher weights to them. But an even better approach might be to estimate these weights more intelligently by using another layer of learning algorithm. This approach is called model stacking.

<p align="center"> 
    <img src="https://i.ibb.co/3RPp594/modelstacking.png">
</p>

Model stacking is an efficient ensemble method in which the predictions, generated by using various machine learning algorithms, are used as inputs in a second-layer learning algorithm. This second-layer algorithm is trained to optimally combine the model predictions to form a new set of predictions. For example, when linear regression is used as second-layer modeling, it estimates these weights by minimizing the least square errors. However, the second-layer modeling is not restricted to only linear models; the relationship between the predictors can be more complex, opening the door to employing other machine learning algorithms.

Lets define a class supporting ensemble learning by stacking multiple models.

````python
class StackingAveragedModels(BaseEstimator, RegressorMixin, TransformerMixin):
    """The class representing a stacking model from a set of models."""
    def __init__(self, base_models, meta_model, n_folds=5):
        """Initialization.
        
        Parameters
        ----------
        base_models: list 
            List of models used to build a averaging model.
        meta_model: ML model
            The meta model (main model) used together with the base models.
        """
        self.base_models = base_models
        self.meta_model = meta_model
        self.n_folds = n_folds
   
    def fit(self, X, y):
        """Fit the data on clones of the original models.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        """
        self.base_models_ = [list() for x in self.base_models]
        self.meta_model_ = clone(self.meta_model)
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=156)
        
        # Train cloned base models then create out-of-fold predictions
        # that are needed to train the cloned meta-model
        out_of_fold_predictions = np.zeros((X.shape[0], len(self.base_models)))
        for i, model in enumerate(self.base_models):
            for train_index, holdout_index in kfold.split(X, y):
                instance = clone(model)
                self.base_models_[i].append(instance)
                instance.fit(X[train_index], y[train_index])
                y_pred = instance.predict(X[holdout_index])
                out_of_fold_predictions[holdout_index, i] = y_pred
                
        # Now train the cloned  meta-model using the out-of-fold predictions as new feature
        self.meta_model_.fit(out_of_fold_predictions, y)
        return self
   
    def predict(self, X):
        """Do the predictions of all base models on the test data.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the testing data
            
        Returns:
        predictions: 1D numpy array
            The predictions by the averaged model.
        """
        # Use the averaged predictions as meta-features for the final prediction 
        # which is done by the meta-model
        meta_features = np.column_stack([
            np.column_stack([model.predict(X) for model in base_models]).mean(axis=1)
                                              for base_models in self.base_models_ ])
        predictions = self.meta_model_.predict(meta_features)
        return predictions
    
    def get_cv_score(self, X, y, cv=10):
        """Compute the k-fold validation on a given data.
        
        Parameters
        ----------
        X: pandas data frame
            The features of the training data
        y: pandas series/data frame
            The labels of the training data
        cv: int
            The number of folds used to test
        
        Returns
        scoreMean: float
            The average of mean-squared errors
        scoreStd: float
            The standard deviation of mean-squared errors
        -------
        """
        scores = np.sqrt(-cross_val_score(self, X, y, scoring='neg_mean_squared_error', cv=cv))
        return scores.mean(), scores.std()
````

````python
models = [ridgeModel, elasticNetModel, gpModel]
stacked_averaging_models = StackingAveragedModels(models, meta_model=bayesianRidgeModel)
    
# Evaluate the ensemble model using 10-fold cross validation
score_mean, score_std = stacked_averaging_models.get_cv_score(X.values, y.values, 10)
print("RMSE of the stacked averaging model: {:.4f} , std: {:.4f}".format(score_mean, score_std))
````

As can be seen, by using the additional second-layer learning algorithm (meta model), the performance of stacked averaging model is higher than the simple averaged model. Because of comprising from many models, the training time of the stacked averaging model is quite long. However, in practice we can take advantages of distributed system to parallelize training component models.

````python
def make_submit(trainDF, testDF, model, file='submission_tmp.csv'):
    """Create the submission file.
    
    Parameters
    ----------
    trainDF: pandas Data Frame.
        The training set
    testDF: pandas Data Frame.
        The test set
    file: string
        The path to the output file.
    """
    # Create the features set and the correspondings target from training set
    X_train = trainDF.drop(columns=['Id', 'SalePrice']).values
    y_train = trainDF.SalePrice.values
    
    # Get the features set of test set
    X_test = testDF.drop(columns = ['Id']).values
    
    # Predict values
    model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = model.predict(X_test)
    
    # Save the results to dataframe and export in csv format
    submitDF = testDF[['Id']].astype(dtype=int)
    submitDF['SalePrice'] = pd.Series(np.exp(y_pred), index=submitDF.index)
    submitDF.to_csv(file, index = False)
    
    return submitDF
````
